```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline 

# To ignore warnings
import warnings
warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv("AmesHousing.csv")
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
#Drop these 2 columns as they are surplus to requirements and not useful
df=df.drop(columns='PID')
df=df.drop(columns='Order')
```

```python
#print out the columns with the most missing values in descending order
missing_values = df.isnull().sum()
highest_counts = missing_values.sort_values(ascending=False)
print(highest_counts)
```
